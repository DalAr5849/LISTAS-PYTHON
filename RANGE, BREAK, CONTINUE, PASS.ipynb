**BREAK**: PERMITE ALTERAR EL COMPORTAMIENTO DE LOS BUCLES "WHILE" Y "FOR".
PERMITE TERMINAR CON LA EJECUCION DEL BUCLE. SALTA A LA SIGUIENTE LNEA DE CODIGO DESPUS DEL BUCLE.

```python
#BREAK CON WHILE
x = 5
while True: #mientras el n煤mero sea menor o igual a 5 continua el ciclo
  x -= 1 #si cumple se le resta 1
  print(x) #se imprime el resultado
  if x == 0: #cuando x=0 se termina el bucle
    break
print("Fin del bucle")
```

```python
#BREAK CON FOR
for i in range(0, 4): #i: toma los valores desde el 0 hasta el 3, 4 iteraciones en total.
  for j in range(0, 4): #j: dentro de cada vuelta de i, comienza un nuevo bucle con (j= 0,3)
    break #rompe el bucle interno (j) despu茅s de la primera vuelta. Entonces j llega soo a valer 0
  print(i, j)
```

```python
#BREAK CON FOR
var = 'Python'
for letra in var:
  if letra == 'h':
    print("Se encontr贸 la letra h")
    break
    print(letra)
```
